1. Load the dataset into a pandas DataFrame and display the first few rows.

3. Check for missing values in each column and handle them appropriately.




```python
#Finding the missing values
print(data_csv.isnull().sum())

#I'm going to replace these values with the means of the columns
data_csv['R&D Spend'] = data_csv['R&D Spend'].fillna(data_csv['R&D Spend'].mean())
data_csv['Marketing Spend'] = data_csv['Marketing Spend'].fillna(data_csv['Marketing Spend'].mean())

#Checking all missing values have been filled in
print(data_csv.isnull().sum())
```




---

4. Calculate the mean, median, and standard deviation of the 'R&D_Spend' column.



```python
#Mean
print(data_csv['R&D Spend'].mean())
#Median
print(data_csv['R&D Spend'].median())
#Standard Deviation
print(data_csv['R&D Spend'].std())
```

7. Group the data by 'State' and calculate the total 'Profit' for each state.



```python
#data_stategroup = data_csv.groupby('State')['Profit'].sum()*

data_stategroup = data_csv.groupby('State').agg({
                                                  'R&D Spend': 'sum',
                                                  'Administration': 'sum',
                                                  'Marketing Spend': 'sum',
                                                  'Profit': 'sum'
                                                }).reset_index()
data_stategroup

#*I could group by just profit, but I'm grouping them all now for a later question
```

8. Create a bar chart to visualize the total 'Profit' by 'State.'



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(3,3))
bar = sns.barplot(data=data_stategroup,x='State',y='Profit',width=.6)
plt.xticks(rotation=45)
plt.title("Total Profit by State")
plt.show
```

18. Calculate the average profit margin for companies in each 'State' and create a bar chart to display it.



```python
data_avgProfitMargin = data_csv.groupby('State').agg({
                                                'R&D Spend': 'sum',
                                                'Administration': 'sum',
                                                'Marketing Spend': 'sum',
                                                'Profit': 'sum',
                                                'Profit Margin': 'mean'
                                              }).reset_index()
plt.figure(figsize=(3,6))
bar = sns.barplot(data=data_avgProfitMargin,x='State',y='Profit Margin',width=.6)
plt.xticks(rotation=45)
plt.title("Average Profit Margin by State")
plt.show

data_avgProfitMargin
```
