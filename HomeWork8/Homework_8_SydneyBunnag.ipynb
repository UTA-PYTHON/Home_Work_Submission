10. Create a new column 'Total_Spend' that represents the sum of 'R&D_Spend,' 'Administration,' and 'Marketing_Spend' for each company.



```python
data_startups["Total_Spend"] = data_startups["R&D Spend"] + data_startups["Marketing Spend"]
data_startups

```

11. Calculate the average **R&D_Spend**, **Administration**, and **Marketing_Spend** for companies located in each **State**



```python
grouped_data = data_startups.groupby("State")

for state, data in grouped_data:
    rd_spend = data["R&D Spend"].mean()
    admin = data["Administration"].mean()
    marketing_spend = data["Marketing Spend"].mean()

    print(f"State: {state}")
    print(f"Average R&D Spend: {rd_spend.round(2)}")
    print(f"Average Administration: {admin.round(2)}")
    print(f"Average Marketing Spend: {marketing_spend.round(2)}\n\n")

```
