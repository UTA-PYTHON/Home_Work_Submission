4. Calculate the mean, median, and standard deviation of the 'R&D_Spend' column.

```python
mean = Startups_df["R&D Spend"].mean() # calculating mean
median = Startups_df["R&D Spend"].median()  # calculating median
standard_Deviation = Startups_df["R&D Spend"].std() # calculating Standard Deviation

# Printing Results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {standard_Deviation}")
```

5. Filter and create a new DataFrame containing companies located in California

```python
# New Data frame for companies located in CA
companies_In_CA = Startups_df[Startups_df['State'] == 'California']
companies_In_CA
```

6. Sort the dataset by 'Profit' in descending order.

```python
# Sorting data set by descending order
sorted_df = Startups_df.sort_values(by='Profit', ascending = False)
sorted_df
```

7. Group the data by 'State' and calculate the total 'Profit' for each state

```python
Startups_df.columns
```

```python
# grouping state and profit & calculating sum
state_Profit = Startups_df.groupby('State')['Profit'].sum()
state_Profit
```

8. Create a bar chart to visualize the total 'Profit' by 'State.'


```python
state_Profit = Startups_df.groupby('State')['Profit'].sum()

state_Profit.plot(kind='bar', color="pink")
plt.title('Total Profit by State')
plt.xlabel('State')
plt.ylabel('Profit')
plt.show()
```

9. Calculate the correlation between 'R&D_Spend' and 'Profit'

```python
# Calculating correlation
corr = Startups_df['R&D Spend'].corr(Startups_df['Profit'])
corr
```

10. Create a new column 'Total_Spend' that represents the sum of 'R&D_Spend,' 'Administration,' and 'Marketing_Spend' for each company.

```python
# Creating new column that represents the sum
Startups_df['Total_Spend'] = Startups_df['R&D Spend'] + Startups_df['Administration'] + Startups_df['Marketing Spend']
Startups_df
```

11. Calculate the average **R&D_Spend**, **Administration**, and **Marketing_Spend** for companies located in each **State**

```python
avg = Startups_df.groupby('State')[['R&D Spend','Administration','Marketing Spend']].mean()
avg
```

12. Create a histogram to visualize the distribution of 'Profit.'

13. Filter the dataset to include only companies with 'R&D_Spend' greater than a 12000.

14. Calculate the total 'Profit' for companies with 'Marketing_Spend' above a certain threshold.

15. Replace any missing values in the 'Marketing_Spend' column with the mean spending of the dataset.

```python
# replacing missing values for Marketing_Spend column
Startups_df["Marketing Spend"] = Startups_df["Marketing Spend"].fillna(Startups_df['Marketing Spend'].mean())
Startups_df
```

16. Create a new column 'Profit_Margin' that represents the profit margin for each company (Profit / Total_Spend).



```python
# creating new column for Profit Margin & calculating
Startups_df['Profit_Margin'] = Startups_df['Profit'] / Startups_df['Total_Spend']
Startups_df
```

17. Sort the dataset by 'Profit_Margin' in descending order.


```python
# Sorting data set "profit margin"
sorted_PM = Startups_df.sort_values(by='Profit_Margin', ascending = False)
sorted_PM
```

18. Calculate the average profit margin for companies in each 'State' and create a bar chart to display it.



```python
# Calculating avg
avg_profit_margin = Startups_df.groupby('State')['Profit_Margin'].mean()
# specifying bar chart and color
avg_profit_margin.plot(kind = 'bar', color = 'purple')
# assigning titles for chart
plt.xlabel('State')
plt.ylabel('Average')
plt.title('Average Profit Margin')
plt.show()
```
