3. Check for missing values in each column and handle them appropriately.




```python
# set 0 values to be NA values
start = start.mask(start == 0, pd.NA)
```

```python
# check for missing values
start.isnull().sum()
```

```python
# handle NA values by foward filling
start.fillna(method = "ffill")
start.head()
```

4. Calculate the mean, median, and standard deviation of the 'R&D_Spend' column.



```python
# describe shows mean and std
print(f"describe data with emphasis on mean and std:\n{start['R&D Spend'].describe()}\n")

# another way to call in mean and std w/ an additional display of mean
print(f"display median of R&D Spend:\n{start['R&D Spend'].median()}\n\ndisplay mean of R&D Spend:\n{start['R&D Spend'].mean()}\n")
print(f"display std of R&D Spend:\n{start['R&D Spend'].std()}")
```

7. Group the data by 'State' and calculate the total 'Profit' for each state.



```python
profit_by_state = start.groupby("State")["Profit"].sum().reset_index(name = "Total")
profit_by_state
```

8. Create a bar chart to visualize the total 'Profit' by 'State.'



```python
sns.barplot(x = "State", y = "Total", data = profit_by_state)
plt.title("Profit by State")
plt.show()
```

10. Create a new column 'Total_Spend' that represents the sum of 'R&D_Spend,' 'Administration,' and 'Marketing_Spend' for each company.



```python
start["Total Spend"] = start["R&D Spend"] + start["Administration"] + start["Marketing Spend"]
start.head(5)
```

11. Calculate the average **R&D_Spend**, **Administration**, and **Marketing_Spend** for companies located in each **State**



```python
avg_by_state = start.groupby("State").agg({"R&D Spend": "mean",
                                           "Administration": "mean",
                                           "Marketing Spend": "mean"}).reset_index()
avg_by_state
```

18. Calculate the average profit margin for companies in each 'State' and create a bar chart to display it.
