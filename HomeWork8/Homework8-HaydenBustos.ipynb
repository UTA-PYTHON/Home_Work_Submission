7. Group the data by 'State' and calculate the total 'Profit' for each state.



```python
state_profits = startup.groupby("State")["Profit"].sum().reset_index()
print(state_profits)
```

8. Create a bar chart to visualize the total 'Profit' by 'State.'



```python
import matplotlib.pyplot as plt

x = state_profits["State"]
y = state_profits["Profit"]
colors = ["purple", "green", "pink"]
plt.bar(x, y, color = colors)
plt.xlabel("State")
plt.ylabel("Profits")
plt.title("Profits by State")
plt.ticklabel_format(style='plain', axis='y')
plt.savefig('profits_by_state.png', dpi=400)
```

10. Create a new column 'Total_Spend' that represents the sum of 'R&D_Spend,' 'Administration,' and 'Marketing_Spend' for each company.



```python
startup["Total_Spend"] = startup["R&D Spend"] + startup["Marketing Spend"]
startup
```

11. Calculate the average **R&D_Spend**, **Administration**, and **Marketing_Spend** for companies located in each **State**



```python
grouped_data = startup.groupby("State")

for state, data in grouped_data:
    rd_spend = data["R&D Spend"].mean()
    admin = data["Administration"].mean()
    marketing_spend = data["Marketing Spend"].mean()

    print(f"State: {state}")
    print(f"Average R&D Spend: {rd_spend.round(2)}")
    print(f"Average Administration: {admin.round(2)}")
    print(f"Average Marketing Spend: {marketing_spend.round(2)}")
    print("--------------------")
```

18. Calculate the average profit margin for companies in each 'State' and create a bar chart to display it.



```python
average_margin = startup.groupby("State")["Profit Margin"].mean().reset_index()

plt.bar(average_margin["State"], average_margin["Profit Margin"])
plt.xlabel("State")
plt.ylabel("Average Profit Margin (%)")
plt.title("Average Profit Margin by State")
plt.show()
```
