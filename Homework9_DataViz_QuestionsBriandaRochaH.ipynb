### **Question 1**

#### Histogram of Median Housing Prices:



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('housing_california.csv')
data

data['median_house_value'].plot(kind='hist')
plt.title('Median Home Value Distribution')
plt.xlabel('median_house_value')
plt.ylabel('Frequency')
plt.show()





```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
data.plot(x='population', y='median_house_value', kind='scatter')
plt.title('Population vs. Median Home Value')
plt.xlabel('Population')
plt.ylabel('Median Home Value')
plt.show()








```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
count = data['ocean_proximity'].value_counts()

count.plot(kind='bar', color='grey')
plt.title('Bar Plot for Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.xticks(rotation=45, size=8)
plt.yticks(rotation=45, size=8)
plt.show()







```

### **Question 6**

#### Correlation Heatmap:


```python
selected_columns = ['median_income', 'median_house_value', 'population', 'housing_median_age']

print(data[selected_columns].head())

correlation_matrix = data[selected_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap for Selected Columns')
plt.show()











```

### **Question 7**

#### Density Plot of Housing Median Age:


```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data['housing_median_age'], fill=True, color='skyblue')
plt.title('Density Plot: Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Density')
plt.show()













```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
data_stacked = data[['total_rooms', 'total_bedrooms']]


plt.figure(figsize=(10, 6))
plt.stackplot(data.index, data_stacked['total_rooms'], data_stacked['total_bedrooms'], labels=['Total Rooms', 'Total Bedrooms'], alpha=0.7)
plt.title('Stacked Area Plot: Total Rooms and Bedrooms')
plt.xlabel('Index')
plt.ylabel('Count')
plt.legend(loc='upper left')
plt.show()
















```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.figure(figsize=(10, 8))
plt.hexbin(data['longitude'], data['latitude'], gridsize=50, cmap='viridis')
plt.title('Hexbin Plot: Latitude vs. Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Count')
plt.show()












```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
selected_columns = ['median_income', 'median_house_value', 'population']

sns.set(style="ticks")
grid = sns.FacetGrid(data, col="ocean_proximity", hue="ocean_proximity", col_wrap=5, height=6)
grid.map(plt.scatter, "median_income", "median_house_value", alpha=0.5)
grid.add_legend()
plt.suptitle('Facet Grid of Scatter Plots: Median Income vs. Median House Value')
plt.show()











```
