```python
# 1. Design a class hierarchy to represent different types of vehicles in a transportation system. Start with a base class called Vehicle with the following attributes and methods:
class Vehicle:
  def __init__(self, make, model, year, fuel):
    self.make=make
    self.model=model
    self.year=year
    self.fuel=fuel
  def summarize(self):
    summary=f"The make is: {self.make}. The model is: {self.model}. The year is: {self.year}. The fuel is: {self.fuel}."
    return summary
```

```python
class Car(Vehicle):
  def __init__(self, make, model, year, fuel):
    super().__init__(make,model,year,fuel)
  def display_Car_info(self):
    "The make is: {self.make}. The model is: {self.model}. The year is: {self.year}. The fuel is: {self.fuel}."
```

```python
class Truck(Vehicle):
  def __init__(self, make, model, year, fuel):
    super().__init__(make,model,year,fuel)
    return
  def display_Truck_info(self):
    "The make is: {self.make}. The model is: {self.model}. The year is: {self.year}. The fuel is {self.fuel}."
```

```python
Car_stuff=Car("Toyota", "Camry", 2016, 10.0)
Truck_stuff = Truck("Ford", "F-150", 2023, 9.0)
```

```python
print(Car_stuff.summarize())
```

```python
print(Truck_stuff.summarize())
```

```python
# 2. Create a subclass called Car with the following additional attributes and methods:
class Car(Vehicle):
  def __init__(self, make, model, year, fuel, num_doors, is_sedan):
    super().__init__(make,model,year,fuel)
    self.num_doors=num_doors
    self.is_sedan=is_sedan
  def Car_info(self):
    display_Car_info=f"The make is: {self.make}. The model is: {self.model}. The year is: {self.year}. The fuel is: {self.fuel}. The number of doors is {self.num_doors}. But is it a sedan? {self.is_sedan}"
    return display_Car_info
```

```python
next_Car=Car("Acura", "MDX", 2008, 17, 4, False)
```

```python
print(next_Car.Car_info())
```

```python
# 3. Create a subclass called Motorcycle with the following additional attributes and methods:
class motorcycle(Vehicle):
  def __init__(self, make, model, year, fuel, top_speed, engine_size):
    super().__init__(make,model,year,fuel)
    self.top_speed=top_speed
    self.engine_size=engine_size
  def motorcycle_info(self):
    display_motorcycle_info=f"The make is: {self.make}. The model is: {self.model}. The year is: {self.year}. The fuel is: {self.fuel}. The top speed is {self.top_speed}. The engine size is {self.engine_size}."
    return display_motorcycle_info
```

```python
Motorcycle=motorcycle("Harley-Davidson", "Trike", 2023, 45.0, 140, 1868.0)
```

```python
print(Motorcycle.motorcycle_info())
```
