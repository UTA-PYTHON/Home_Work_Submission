### **Question 1**

#### Histogram of Median Housing Prices:



```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('housing_california.csv')


plt.figure(figsize=(8, 6))
plt.hist(data['median_house_value'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Median Housing Prices')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('housing_california.csv')


plt.figure(figsize=(8, 6))
plt.scatter(data['population'], data['median_house_value'], alpha=0.5, c='orange')
plt.title('Scatter Plot of Population vs. Median House Value')
plt.xlabel('Population')
plt.ylabel('Median House Value')
plt.grid(True)
plt.show()

```

### **Question 4**

#### Pair Plot for Select Columns:


```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('housing_california.csv')


selected_columns = ['median_house_value', 'median_income', 'population', 'total_bedrooms']

sns.pairplot(data[selected_columns])
plt.suptitle('Pair Plot of Selected Columns', y=1.02)
plt.show()
```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('housing_california.csv')

ocean_proximity_counts = data['ocean_proximity'].value_counts()


plt.figure(figsize=(8, 6))
ocean_proximity_counts.plot(kind='bar', color='skyblue')
plt.title('Counts of Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

### **Question 6**

#### Correlation Heatmap:


```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('housing_california.csv')

# Calculate correlations
correlation_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

```

### **Question 7**

#### Density Plot of Housing Median Age:


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('housing_california.csv')


plt.figure(figsize=(8, 6))
sns.histplot(data['housing_median_age'], kde=True, color='purple')
plt.title('Density Plot of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Density')
plt.grid(True)
plt.show()

```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('housing_california.csv')

data = data.sort_values('housing_median_age')

plt.figure(figsize=(10, 6))

plt.stackplot(data['housing_median_age'], data['total_rooms'], data['total_bedrooms'],
              labels=['Total Rooms', 'Total Bedrooms'], colors=['lightblue', 'salmon'])

plt.title('Stacked Area Plot of Total Rooms and Bedrooms')
plt.xlabel('Housing Median Age')
plt.ylabel('Count')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()

```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('housing_california.csv')

plt.figure(figsize=(8, 6))
plt.hexbin(data['longitude'], data['latitude'], gridsize=100, cmap='YlGnBu')
plt.colorbar(label='Count in Bin')
plt.title('Hexbin Plot of Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

### **Question 10**
