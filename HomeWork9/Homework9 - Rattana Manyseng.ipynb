```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("housing_california.csv")
df.head()
```

## 1.Histogram of Median Housing Prices:

```python
df['median_house_value'].plot(kind='hist',color="blue",edgecolor="blue")
plt.title('Median Housing Prices')
plt.xlabel('Value')
plt.ylabel('Median Estimation')
plt.show()
```

## 2. Scatter Plot of Population vs. Median House Value:

```python
df.plot(x='median_house_value',y='population',kind='scatter')
plt.title('Population vs. Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Population')
plt.show()
```

## 3. Box Plot of Median Income:

```python
plt.boxplot(df['median_income'])
plt.xlabel("Median Income")
plt.ylabel('Value')
plt.title('Box Plot of Median Income')
plt.show()
```

## 8. Stacked Area Plot of Total Rooms and Bedrooms:


```python
df[['total_rooms','total_bedrooms']].plot(kind='area',stacked=True)
plt.xlabel("Total Number Rooms/Bedrooms")
plt.ylabel("Volume")
plt.title("Stacked Area Plot Rooms/Bedrooms")
plt.show()
```

## 9. Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(df['latitude'],df['longitude'],gridsize=8,cmap='BuPu')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Hexbin Plot')
plt.show()
```

## 10. Facet Grid of Scatter Plots:
