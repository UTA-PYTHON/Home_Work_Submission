### Use the `housing_california.csv` dataset to answer the questions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("housing_california.csv")
```

### **Question 1**

#### Histogram of Median Housing Prices:



```python
sns.histplot(df["median_house_value"])
plt.ylabel("Frequency")
plt.xlabel("Median Housing Prices")
plt.title("Histogram for Median Housing Prices")
plt.show()
```



---

---



### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
plt.scatter(df["population"],df['median_house_value'], alpha=0.2)
plt.xlabel("Population")
plt.ylabel("Median House Value")
plt.title("Population vs Median House Value")
plt.show()
```

### **Question 3**

#### Box Plot of Median Income:



```python
sns.boxplot(x="median_income", data = df)
plt.xlabel("Median Income")
plt.title("Box Plot of Median Income")
plt.show()
```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
df["ocean_proximity"].value_counts().plot(kind="bar")
plt.xlabel("Ocean Proximity")
plt.ylabel("Frequency")
plt.title("Bar Plot of Ocean Proximity Counts")
plt.show()
```

### **Question 6**

#### Correlation Heatmap:


```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="viridis",linewidth=0.5,fmt=".2f",annot_kws={"size":12})
plt.title("Correlation Heatmap")
```

### **Question 7**

#### Density Plot of Housing Median Age:

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
df[["total_rooms","total_bedrooms"]].plot(kind="area",stacked=True)
plt.xlabel("Sample Index")
plt.ylabel("Count")
plt.title("Stacked Area Plot of Total Rooms and Bedrooms")

plt.show()
```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(df["latitude"],df["longitude"],gridsize=10,cmap="PiYG_r")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Hexbing Plot of Latitude and Longitude")
plt.colorbar()
plt.show()
```
