### **Question 1**

#### Histogram of Median Housing Prices:



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("housing_california.csv")
print(df.head()) # printing this is to keep track of the fields

df['median_house_value'].plot(kind='hist', bins=10, edgecolor='black')
plt.xlabel('Median House Values')
plt.title('Median Housing Prices')
plt.show()
```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
plt.scatter(df['population'], df['median_house_value'])
plt.xlabel('Population')
plt.ylabel('Median House Value')
plt.title('Scatter Plot of Population vs. Median House Value')
plt.show()
```

### **Question 4**

#### Pair Plot for Select Columns:

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
df['ocean_proximity'].value_counts().plot(kind='bar', color='red', edgecolor='black')
plt.title('Ocean Proximity Counts')
plt.xlabel('Ocean Proximities')
plt.ylabel('Counts')
plt.show()
```

### **Question 6**

#### Correlation Heatmap:


```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', lw = 0.5, fmt='.2f',
            annot_kws = {"size": 12})
```

### **Question 7**

#### Density Plot of Housing Median Age:


```python
sns.kdeplot(df['housing_median_age'], fill=True, color='green')
plt.title('Housing Median Age')
plt.show()
```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
df[['total_rooms', 'total_bedrooms']].plot(kind='area', stacked=True)
plt.title('Total Rooms and Bedrooms')
plt.xlabel('Number of Respective Rooms')
plt.ylabel('Count')
plt.show()
```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(df['longitude'], df['latitude'], gridsize=20, cmap='Purples')
plt.title('Latitude and Longitude')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.colorbar(label='Count')
plt.show()
```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
grid = sns.FacetGrid(df, col='ocean_proximity', col_wrap=3, height=4)
grid.map(plt.scatter, 'median_house_value', 'population', alpha=0.5).add_legend()
grid.set_axis_labels("Median House Value", 'Population')
grid.fig.suptitle('Facet Grid of Scatter Plots of Median House Value and Population', y=1.02)
plt.show()
```
