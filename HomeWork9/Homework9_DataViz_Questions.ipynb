### **Question 1**

#### Histogram of Median Housing Prices:



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

!ls
sub_housing_california_df=pd.read_csv("housing_california.csv")
sub_housing_california_df.head()

df=pd.DataFrame(sub_housing_california_df)

# Plotting a histogram of median house values
plt.figure(figsize=(10, 6))
plt.hist(df['median_house_value'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Median Housing Prices')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()






```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
# Scatter plot of Population vs. Median House Value
plt.figure(figsize=(10, 6))
plt.scatter(df['population'], df['median_house_value'], alpha=0.5, color='green')
plt.title('Scatter Plot: Population vs. Median House Value')
plt.xlabel('Population')
plt.ylabel('Median House Value')
plt.grid(True)
plt.show()









```

### **Question 3**

#### Box Plot of Median Income:



```python
# Box plot of Median Income
plt.figure(figsize=(10, 6))
plt.boxplot(df['median_income'], vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Box Plot of Median Income')
plt.xlabel('Median Income')
plt.grid(True)
plt.show()










```

### **Question 4**

#### Pair Plot for Select Columns:


```python

# Selecting columns for the pair plot
selected_columns = ['population', 'median_income', 'median_house_value', 'total_rooms']

# Creating a DataFrame with the selected columns
selected_df = df[selected_columns]

# Pair plot for selected columns
sns.pairplot(selected_df)
plt.suptitle('Pair Plot of Selected Columns', y=1.02)
plt.show()









```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
# Bar plot of Ocean Proximity Counts
plt.figure(figsize=(10, 6))
sns.countplot(x='ocean_proximity', data=df, palette='viridis')
plt.title('Bar Plot of Ocean Proximity Counts')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.show()








```



---

### **Question 6**

#### Correlation Heatmap:


```python
# Selecting numeric columns for the correlation heatmap
numeric_columns = ['population', 'median_income', 'median_house_value', 'total_rooms']

# Creating a DataFrame with the selected columns
numeric_df = df[numeric_columns]

# Creating a correlation matrix
correlation_matrix = numeric_df.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Selected Numeric Columns')
plt.show()











```

### **Question 7**

#### Density Plot of Housing Median Age:


```python
# Density plot of Housing Median Age
plt.figure(figsize=(10, 6))
sns.kdeplot(df['housing_median_age'], fill=True, color='skyblue')
plt.title('Density Plot of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Density')
plt.show()













```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
# Creating a stacked area plot for 'total_rooms' and 'total_bedrooms'
plt.figure(figsize=(12, 8))

plt.stackplot(df.index, df['total_rooms'], df['total_bedrooms'], labels=['Total Rooms', 'Total Bedrooms'], alpha=0.7)
plt.legend(loc='upper left')
plt.title('Stacked Area Plot of Total Rooms and Bedrooms')
plt.xlabel('Index')
plt.ylabel('Count')
plt.show()















```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
# Hexbin plot of Latitude and Longitude
plt.figure(figsize=(10, 8))
plt.hexbin(df['longitude'], df['latitude'], gridsize=50, cmap='viridis')
plt.colorbar(label='Count')
plt.title('Hexbin Plot of Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()












```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
# Creating a facet grid of scatter plots
scatter_grid = sns.FacetGrid(df, col="ocean_proximity", col_wrap=3, height=4)
scatter_grid.map(plt.scatter, "longitude", "latitude", alpha=0.5)
scatter_grid.set_axis_labels("Longitude", "Latitude")
scatter_grid.set_titles(col_template="{col_name}")
plt.suptitle('Facet Grid of Scatter Plots by Ocean Proximity', y=1.02)
plt.show()











```
