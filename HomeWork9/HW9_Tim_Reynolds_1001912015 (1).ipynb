### **Question 1**

#### Histogram of Median Housing Prices:



```python
data = pd.read_csv("housing_california.csv")
sns.histplot(data=data['median_house_value'])
```

### **Question 4**

#### Pair Plot for Select Columns:


```python
data_cols = ['median_house_value', 'median_income', 'population']
data_selected = data[data_cols]
sns.set_context('talk', font_scale=2)
sns.pairplot(data = data_selected, markers=['o','s','D'], height=15, plot_kws={'s':50})
plt.subplots_adjust(top=0.30)
```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
plt.figure(figsize=(12, 6))
sns.set_context('talk', font_scale=1)
sns.countplot(data=data, x='ocean_proximity', palette='viridis')
plt.xlabel('Proximity')
plt.ylabel('Count')
plt.title('Count of Ocean Proximity')
plt.show()
```

### **Question 6**

#### Correlation Heatmap:


```python
correlation_matrix=data.corr()

sns.heatmap(correlation_matrix, annot=True,cmap='viridis',linewidths=0.5,
            fmt='.2f', annot_kws={'size':12})
```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
sns.lineplot(data=data[['total_rooms', 'total_bedrooms']],
             palette="dark:salmon_r", linewidth=2,alpha=0.8)
plt.ylabel('Amount')
plt.show()
```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
x=data['latitude']
y=data['longitude']

plt.hexbin(x,y,gridsize=10, cmap="CMRmap", alpha=0.5)

plt.xlabel('Lat')
plt.ylabel('Lon')

plt.xlim(32.5,40)
plt.ylim(-124,-116)
plt.colorbar()
```
