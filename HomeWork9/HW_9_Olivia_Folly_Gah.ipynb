```python
sub_housing = "housing_california.csv"
sub_housing_df = pd.read_csv(sub_housing)
```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
plt.scatter(sub_housing_df["population"], sub_housing_df["median_house_value"], color="blue")
plt.title("Population vs. Median House Value")
plt.show()
```

### **Question 3**

#### Box Plot of Median Income:



```python
plt.boxplot(sub_housing_df["median_income"])
plt.title("Box Plot of the Median Income")
plt.show()
```

### **Question 4**

#### Pair Plot for Select Columns:


```python
select_columns=["housing_median_age", "median_income"]
sns.pairplot(sub_housing_df[select_columns])
plt.suptitle('Pair Plot of Median Income and Median Age', y=1.05)
plt.show()










```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
ocean_prox=sub_housing_df["ocean_proximity"].value_counts()
ocean_prox.plot(kind='bar', color="#8899AC", edgecolor='b')
plt.title("Bar Plot of Ocean Proximity")
plt.xlabel("Ocean Proximity")
plt.ylabel("Counts")
plt.show()
```



---



---



---



### **Question 6**

#### Correlation Heatmap:


```python
correlation_matrix=sub_housing_df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap="viridis", linewidths=0.5,
            fmt=".2f", annot_kws={"size":12})
```



---



---



---



### **Question 7**

#### Density Plot of Housing Median Age:


```python
sns.kdeplot(sub_housing_df["housing_median_age"], fill=True, color='#FFA500')
plt.title("Density Plot of Housing Median Age")
plt.xlabel("Median Age")
plt.ylabel('Density')
plt.show()
```



---



---

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
rooms=["total_rooms", "total_bedrooms"]
colors=["purple", "teal"]
sub_housing_df[rooms].plot(kind='area', stacked=True, alpha=1, color=colors)
plt.title('Stacked Area Plot of Total Rooms and Total Bedrooms')
plt.ylabel('Count')
plt.show()
```



---



---



---



### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(sub_housing_df["longitude"], sub_housing_df["latitude"], gridsize=45, cmap='viridis', bins='log')
plt.title("Longitude vs Latittude")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
columns = ["ocean_proximity", "median_income", "median_age", "total_rooms", "total_bedrooms", "population", "households", "median_house_value"]
g = sns.FacetGrid(sub_housing_df, col="ocean_proximity", hue="ocean_proximity", col_wrap=3, height=5)
g.map(plt.scatter, "longitude", "latitude", alpha=0.5)
g.add_legend()
plt.show()
```
