# **Week 13. Data Visualization Homework**

### Use the `housing_california.csv` dataset to answer the questions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("housing_california.csv")

data.columns
```

### **Question 1**

#### Histogram of Median Housing Prices:



```python
plt.hist(data["median_house_value"],
        bins=80)

plt.xlabel("Median Housing Prices")
plt.ylabel("Number of Locations")
plt.title("Median Housing Price Distribution")

plt.show()
```



---



---



---



### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
plt.scatter(data["median_house_value"],data["population"])

plt.xlabel("Median Housing Prices")
plt.ylabel("Population")
plt.title("Population vs. Median House Value")

plt.show()
```

### **Question 3**

#### Box Plot of Median Income:



```python
plt.figure(figsize=(10,1))
plt.grid()

sns.boxplot(x="median_income",
            data=data)
plt.xlabel("Median Income (per hr)")
plt.title("Box Plot Median Income")

plt.show()
```

### **Question 4**

#### Pair Plot for Select Columns:


```python
# using fewer columns because my laptop keeps crashing when I use all of the columns
data_condensed=pd.read_csv("housing_california.csv",
                        usecols=['longitude',
                                 'latitude',
                                 'population',
                                 'households',
                                 'median_income',
                                 'median_house_value',
                                 'ocean_proximity'])
sns.pairplot(data_condensed,
             hue="ocean_proximity")

plt.show()
```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
sns.countplot(x="ocean_proximity",
              data=data)

plt.ylabel("Count")
plt.xlabel("Ocean Proximity")
plt.title("Count of the Houses' Proximity to the Ocean")

plt.show()
```



---

### **Question 6**

#### Correlation Heatmap:


```python
data_noocean=pd.read_csv("housing_california.csv",
                        usecols=['longitude',
                                 'latitude',
                                 'housing_median_age',
                                 'total_rooms',
                                 'total_bedrooms',
                                 'population',
                                 'households',
                                 'median_income',
                                 'median_house_value'])

correlation=data_noocean.corr()

sns.heatmap(correlation,
            annot=True,
            cmap="twilight")

plt.title("Correlation Heatmap of California Housing")

plt.show()
```

### **Question 7**

#### Density Plot of Housing Median Age:


```python
sns.kdeplot(data["housing_median_age"], 
            fill=True)

plt.xlabel("Median Housing Age")
plt.title("Housing Median Age Density Plot")

plt.show()
```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
data[["total_rooms","total_bedrooms"]].plot(kind="area",stacked=True)

plt.xlabel("Total Bedrooms or Rooms")
plt.ylabel("Count")
plt.title("Stacked Area Plot: Total Rooms and Total Bedrooms")
plt.legend(loc="upper right")

plt.show()
```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(data["latitude"],
           data["longitude"],
          gridsize=100,
          cmap="twilight")

plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Hexbin Plot: Longitude vs Latitude")

plt.show()
```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
f=sns.FacetGrid(data,
                col="ocean_proximity")
f.map_dataframe(sns.scatterplot,
                "median_income",
                "median_house_value",
                alpha=0.2)

plt.show()
```
