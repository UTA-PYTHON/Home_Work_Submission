### **Question 1**

#### Histogram of Median Housing Prices:



```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('/content/sample_data/housing_california.csv')

plt.hist(data['median_house_value'],bins=30,color='skyblue')
plt.xlabel('Median House Value')
plt.ylabel('Histogram of Median Housing Prices')
plt.show()
```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
plt.scatter(data['population'], data['median_house_value'], alpha=0.5, color='green')
plt.xlabel('Population')
plt.ylabel('Median House Value')
plt.title('Scatter Plot of Population vs. Median House Value')
plt.show()
```

### **Question 3**

#### Box Plot of Median Income:



```python
plt.boxplot(data['median_income'])
plt.xlabel('Median Income')
plt.title('Box Plot of Median Income')
plt.show()
```

### **Question 4**

#### Pair Plot for Select Columns:


```python
import seaborn as sns

selected_columns = ['median_house_value', 'median_income', 'total_rooms', 'housing_median_age']

sns.pairplot(data[selected_columns])
plt.title('Pair Plot of Selected Columns')
plt.show()
```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
ocean_counts = data['ocean_proximity'].value_counts()
ocean_counts.plot(kind='bar', color='orange')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.title('Bar Plot of Ocean Proximity Counts')
plt.show()
```



---



---



---



### **Question 6**

#### Correlation Heatmap:


```python
numeric_columns = data.select_dtypes(include='number')

corr_matrix = numeric_columns.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```



---



---

### **Question 7**

#### Density Plot of Housing Median Age:


```python
sns.kdeplot(data['housing_median_age'], fill=True, color='purple')
plt.xlabel('Housing Median Age')
plt.ylabel('Density')
plt.title('Density Plot of Housing Median Age')
plt.show()
```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
data[['total_rooms', 'total_bedrooms']].plot(kind='area', stacked=True)
plt.xlabel('Sample Index')
plt.ylabel('Count')
plt.title('Stacked Area Plot of Total Rooms and Bedrooms')
plt.show()
```



---



---



---



### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(data['longitude'], data['latitude'], gridsize=50, cmap='viridis')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Hexbin Plot of Latitude and Longitude')
plt.colorbar(label='Count in bin')
plt.show()
```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
g = sns.FacetGrid(data, col='ocean_proximity', col_wrap=3)
g.map(plt.scatter, 'median_house_value', 'median_income', alpha=0.5)
g.set_axis_labels('Median House Value', 'Median Income')
g.fig.suptitle('Facet Grid of Scatter Plots by Ocean Proximity', y=1.02)
plt.show()
```
