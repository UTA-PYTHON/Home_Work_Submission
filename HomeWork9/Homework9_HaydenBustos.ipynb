# **Week 13. Data Visualization Homework**

### Use the `housing_california.csv` dataset to answer the questions.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

### **Question 1**

#### Histogram of Median Housing Prices:



```python
house = pd.read_csv("housing_california.csv")
plt.hist(house["median_house_value"], bins = 20)
plt.title("Median House Value")
plt.ylabel("Frequency")
plt.xlabel("Cost ($)")
```



---



---



---



### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
population = house["population"]
house_value = house["median_house_value"]
plt.scatter(population, house_value, alpha = .25)
plt.xlabel("Population")
plt.ylabel("Median House Value")
plt.title("Population vs. Median House Value")
```



---

### **Question 3**

#### Box Plot of Median Income:



```python
plt.figure(figsize= (12,4))
plt.boxplot(house['median_income'], vert = False)
plt.xlabel("Median Income")
plt.ylabel("Box Plot")
```



---



---



---



### **Question 4**

#### Pair Plot for Select Columns:


```python
selected_columns = ['median_house_value', 'total_bedrooms', "median_income", "housing_median_age"]
pairplot_c = house[selected_columns]
sns.pairplot(pairplot_c)
```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
house["ocean_proximity"].value_counts().plot(kind = "bar")
plt.xticks(rotation = 0)
plt.ylabel("Frequency", fontsize = 14)
plt.xlabel("Ocean Proximity", fontsize = 14)
plt.title("Ocean Proximity Frequency")
```



---



---



---



### **Question 6**

#### Correlation Heatmap:


```python
correlation_matrix = house.corr()
plt.figure(figsize=(16, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.xticks(rotation = 45)
plt.title('Correlation Heatmap')
plt.show()
```



---



---



---



### **Question 7**

#### Density Plot of Housing Median Age:


```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=house["housing_median_age"], shade = True)
plt.xlabel("Housing Median Age", fontsize = 14)
plt.title("Housing Median Age Density")
plt.show()
```



---



---



---



### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
plt.figure(figsize=(10, 6))

plt.stackplot( house['total_rooms'], house['total_bedrooms'], labels=['Total Rooms', 'Bedrooms'], alpha=0.5)
plt.xlabel("Total Rooms", fontsize = 14)
plt.ylabel("Total Bedrooms", fontsize = 14)
plt.title("Bedrooms and Total Rooms")
plt.show()
```



---



---



---



### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.figure(figsize=(14, 8))
plt.hexbin(x = house["longitude"], y = house["latitude"], gridsize=100, cmap="cividis")

plt.colorbar(label='Density')
plt.xlabel("Longitude", fontsize = 14)
plt.ylabel("Latitude", fontsize = 14)
plt.title("Longitude vs Latitude")
plt.show()
```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
g = sns.FacetGrid(house, col="ocean_proximity", col_wrap=3, height=4)
g.map(plt.scatter, "median_income", "median_house_value", alpha=0.5)
g.set_axis_labels("Median Income", "Median House Value")

plt.show()
```
