### **Question 1**

#### Histogram of Median Housing Prices:



```python
H9_Med_House= H9_read['median_house_value']

plt.hist(H9_Med_House, bins=100, color= 'pink')
plt.xlabel('Housing Price')
plt.ylabel('Amount of Homes')
plt.title("Median Housing Prices")











```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
H9_population= H9_read['population']
H9_Median_House_Value=H9_read['median_house_value']

plt.scatter(H9_population, H9_Median_House_Value, color='pink' )
plt.title('Scatter Plot of Population vs. Median House Value')
plt.xlabel('Population in California')
plt.ylabel('Median House Value')








```
