# **Week 13. Data Visualization Homework**

### Use the `housing_california.csv` dataset to answer the questions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("housing_california.csv")
df
```

### **Question 1**

#### Histogram of Median Housing Prices:



```python
df['median_house_value'].plot(kind='hist', color="gray", edgecolor="black")
plt.title('Median Housing Prices')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```



---



---



---



### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
df.plot(x='population', y='median_house_value', kind='scatter')
plt.title('Population vs. Median House Value')
plt.xlabel('Population')
plt.ylabel('Median House Value')
plt.show()
```

### **Question 3**

#### Box Plot of Median Income:



```python
plt.boxplot(df['median_income'])
plt.xlabel('Median Income')
plt.ylabel('Value')
plt.title('Box Plot of Median Income')
plt.show()
```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
counts = df['ocean_proximity'].value_counts()

# Create a bar plot
counts.plot(kind='bar', color='skyblue')
plt.title('Bar Plot for Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.xticks(rotation=45, size=8)
plt.yticks(rotation=45, size=8)
plt.show()
```

### **Question 6**

#### Correlation Heatmap:


```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="viridis",linewidth=0.5, fmt=".2f",annot_kws={"size": 12})
```



---

### **Question 7**

#### Density Plot of Housing Median Age:


```python
sns.kdeplot(df['housing_median_age'], fill=True, color='skyblue', edgecolor="lightblue")

plt.title('Density Plot of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.show()
```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
fig, ax = plt.subplots()


ax.stackplot(df['total_rooms'], df['total_bedrooms'], labels=['Total Rooms', 'Bedrooms'], alpha=0.5)
ax.set_title('Stacked Area Plot of Total Rooms and Bedrooms')
ax.set_xlabel('Area')
ax.set_ylabel('Number of Rooms')
ax.legend(loc='upper left')
plt.show()
```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(df['longitude'], df['latitude'], gridsize=10, cmap='PiYG_r')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Hexbin Plot')
plt.colorbar()
plt.show()
```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
fg = sns.FacetGrid(df, col="ocean_proximity", col_wrap=2, height=4)
fg.map(sns.scatterplot, "longitude", "latitude")
fg.set_axis_labels("Longitude", "Latitude")
fg.set_titles(col_template="Ocean Proximity = {col_name}")
plt.show()
```
