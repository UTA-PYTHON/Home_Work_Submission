### **Question 1**

#### Histogram of Median Housing Prices:



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

housing_cali=pd.read_csv("/Users/poiismac/Downloads/housing_california.csv")
```

```python
housing_cali.head(3)
```

```python
#finding any missing value 
housing_cali.isnull().sum()
```

```python

plt.hist(housing_cali["median_house_value"],density=True, label="Median House Price")
plt.xlabel("Median House Price",fontsize=15)
plt.ylabel("Probability",fontsize=15)
plt.legend()


plt.title("Histogram of Median House Value",fontsize=20)
plt.show()
```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
plt.figure(figsize=(25,9))

sns.scatterplot(housing_cali,x="median_house_value",y="population",hue="housing_median_age")

plt.xlabel("Average House Value",fontsize=20)
plt.ylabel("Number of Population",fontsize=20)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)


plt.title("Scatterplot: Number of Population VS Average House Value",fontsize=30)



plt.show()

```

### **Question 3**

#### Box Plot of Median Income:



```python
plt.figure(figsize=(25,9))

sns.boxplot(housing_cali,x="housing_median_age",y="median_income")


plt.xlabel("Average Age",fontsize=20)
plt.ylabel("Average Income",fontsize=20)

plt.title("Median Income at Different Ages",fontsize=20)

plt.show()

```

### **Question 4**

#### Pair Plot for Select Columns:


```python
sns.pairplot(housing_cali,vars=['housing_median_age','population','median_income','median_house_value'])

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()



```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
plt.figure(figsize=(20,15))

sns.boxplot(data=housing_cali,x="ocean_proximity",y='longitude')

plt.xlabel("Ocean Proximity",fontsize=20)
plt.ylabel("Longitude",fontsize=20)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)


plt.title("Bar Plot of Ocean Proximity Counts", fontsize= 20)

plt.show()





```

### **Question 6**

#### Correlation Heatmap:


```python
housing_cali.columns
```

```python
plt.figure(figsize=(20,8))

columns_heatmap=['housing_median_age',
                 'total_rooms','total_bedrooms','population','households','median_income','median_house_value']

heatmap_df=housing_cali[columns_heatmap]
sns.heatmap(heatmap_df.corr(),annot=True,fmt='.2f',cmap='coolwarm')


plt.show()







```
