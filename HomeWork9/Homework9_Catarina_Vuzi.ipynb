# **Week 13. Data Visualization Homework**


### Use the `housing_california.csv` dataset to answer the questions.



```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
!ls
```

```python
df=pd.read_csv("housing_california.csv")
df.head(3)
```


# **Question 1: Histogram of Median Housing Prices**



```python
sns.histplot(df['median_house_value'])
plt.title('Median Housing Price Distribution')
plt.show()
```

# **Question 2: Scatter Plot of Population vs. Median House Value**

```python
plt.scatter(df["population"], df["median_house_value"], alpha=0.5,color="indigo")
plt.show()
```

# **Question 3: Box Plot of Median Income**






```python
df.boxplot(column='median_income')
plt.show()
```

# **Question 4: Pair Plot for Select Columns**

```python

sns.set_context('notebook')
sns.pairplot(df,hue='population')
plt.show()
```

# **Question 5: Bar Plot of Ocean Proximity Counts**

```python
sns.countplot(x='ocean_proximity', data=df)
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.title('Bar Plot of Ocean Proximity Counts')
plt.show()
```

# **Question 6: Correlation Heatmap**

```python

sns.heatmap(df.corr(),cmap='Blues',lw=0.5,annot=True,fmt='.2f',annot_kws={'size':12})
```

# **Question 7: Density Plot of Housing Median Age**





```python
df["housing_median_age"].plot(kind='density')
```

 # **Question 8: Stacked Area Plot of Total Rooms and Bedrooms**





```python
df[['total_rooms','total_bedrooms']].plot(kind='area',stacked=True)
plt.xlabel('Number of Rooms or Bedrooms')
plt.ylabel('Counts')
plt.title('Stacked Area Plot for the Total Rooms and Total Bedrooms')
plt.legend(loc='upper left')
plt.show()

```

# **Question 9: Hexbin Plot of Latitude and Longitude**

```python
plt.hexbin(df['longitude'], df['latitude'], gridsize=20, cmap='Blues')
plt.colorbar()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

# **Question 10: Facet Grid of Scatter Plots**

```python
g=sns.FacetGrid(df, col='ocean_proximity')
g.map(plt.scatter, "median_income", "median_house_value",alpha=0.1).add_legend()
plt.show()
```
