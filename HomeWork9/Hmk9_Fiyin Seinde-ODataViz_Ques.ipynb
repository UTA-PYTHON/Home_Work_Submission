### **Question 1**

#### Histogram of Median Housing Prices:



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset= pd.read_csv("/housing_california.csv")

sns.histplot(dataset['median_house_value'])
plt.title('Median Housing Prices')
plt.show()
```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
plt.scatter(dataset['population'], dataset['median_house_value'], alpha=0.5, color='yellow')
plt.title('Scatter Plot of Population vs Median House Value')
plt.show()
```

### **Question 4**

#### Pair Plot for Select Columns:


```python
selectcolumn= ['median_house_value', 'median_income','total_rooms', 'housing_median_age']

sns.pairplot(dataset[selectcolumn])
plt.title('Pair Plot of Select Column')
plt.show()
```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
oceanprox= dataset['ocean_proximity'].value_counts()
oceanprox.plot(kind='bar', color='yellow')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.title('Bar Plot of Ocean Proximity Counts')
plt.show()

```

### **Question 6**

#### Correlation Heatmap:


```python
numeric_columns= dataset.select_dtypes(include='number')
corr_matrix= numeric_columns.corr()
sns.heatmap(corr_matrix, annot=True, cmap='spring', fmt='0.2f')
plt.title('Correlation Heatmap')
```

### **Question 7**

#### Density Plot of Housing Median Age:


```python
sns.kdeplot(dataset['housing_median_age'], fill=True, color='yellow')
plt.title('Density Plot of Housing Median Age')
plt.xlabel('HMA')
plt.ylabel('Density')
```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
dataset[['total_rooms','total_bedrooms']].plot(kind='area', stacked=True)
plt.title('Stacked Area Plot of Toatl Rooms and Bedrooms')
plt.xlabel('Sample Index')
plt.ylabel('Count')
```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(dataset['longitude'], dataset['latitude'], gridsize=30, cmap='viridis')
plt.title('Hexbin Plot of Lat and Longitude')
plt.colorbar()
plt.xlabel('Longitude')
plt.ylabel('Latitudee')
```
