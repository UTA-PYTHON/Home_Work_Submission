# **Week 13. Data Visualization Homework**

### Use the `housing_california.csv` dataset to answer the questions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("housing_california.csv")
```

### **Question 1**

#### Histogram of Median Housing Prices:



```python
plt.hist(df["median_house_value"],bins=60)
plt.xlabel("Median Housing Prices")
plt.ylabel("Number of Locations")
plt.title("Median Housing Price Distribution")

plt.show()


```



---



---



---



### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
plt.scatter(x=df['population'], y=df['median_house_value'], alpha=0.5)

plt.xlabel('Population')
plt.ylabel('Median House Value($)')
plt.title('Population vs Median House Value')

plt.show()

```



---

### **Question 3**

#### Box Plot of Median Income:



```python
sns.boxplot(df['median_income'])
plt.title('Median Income')
plt.show()




```



---

### **Question 4**

#### Pair Plot for Select Columns:


```python
df2 = df[['median_house_value','median_income']]
sns.pairplot(df2)
plt.show()
```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
df['ocean_proximity'].value_counts().plot(kind='bar',color=['r','b','y','g'])
plt.title('Distribution of Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.show()








```



---



---



---



### **Question 6**

#### Correlation Heatmap:


```python
sns.set_context('poster',font_scale=0.5)
sns.heatmap(df.corr(), cmap='YlGn', annot=True, fmt='.2f', lw=0.5)
plt.title('Correlation Heatmap for California Housing')
plt.show()
```



---



---



---



### **Question 7**

#### Density Plot of Housing Median Age:


```python
sns.kdeplot(df['housing_median_age'])
plt.xlabel('Housing Median Age')
plt.title('Housing Median Age Density Plot')
plt.show()
```



---



---



---



### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
df[['total_rooms','total_bedrooms']].plot(kind='area', stacked=True)
plt.title('Stacked Area Plot of Total Rooms and Bedrooms')
plt.legend(loc='upper left')
plt.show()
```



---



---



---



### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(df['latitude'], df['longitude'])
plt.colorbar()
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Latitude and Longitude Hexbin')
plt.show()
```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
grid=sns.FacetGrid(df, col='ocean_proximity', height=6)
grid.map(sns.scatterplot, 'median_income', 'median_house_value',alpha=0.7)
grid.set_axis_labels('Median Income','Median House Value',fontsize=20)
plt.show()
```
