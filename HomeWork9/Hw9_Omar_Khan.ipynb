### **Question 1**

#### Histogram of Median Housing Prices:



```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
import matplotlib.pyplot as plt

# Load
file_name = "housing_california.csv"
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

print(df.head())

# Plot
plt.hist(df['median_house_value'], bins=50, color='blue', alpha=0.7)
plt.title('Histogram of Median Housing Prices')
plt.xlabel('Median Housing Price')
plt.ylabel('Frequency')
plt.show()
```

### **Question 6**

#### Correlation Heatmap:


```python
import pandas as pd
import io
import seaborn as sns
import matplotlib.pyplot as plt

file_name = "housing_california.csv"
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

# Select columns for the correlation heatmap
selected_columns = ['median_income', 'median_house_value', 'population', 'housing_median_age']

print(df[selected_columns].head())

# Calculate the correlation matrix for the selected columns
correlation_matrix = df[selected_columns].corr()

# Create a correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap for Selected Columns')
plt.show()
```

### **Question 7**

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
import pandas as pd
import io
import matplotlib.pyplot as plt

file_name = "housing_california.csv"
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

print(df.head())

df_stacked = df[['total_rooms', 'total_bedrooms']]

# Create a stacked area plot
plt.figure(figsize=(10, 6))
plt.stackplot(df.index, df_stacked['total_rooms'], df_stacked['total_bedrooms'], labels=['Total Rooms', 'Total Bedrooms'], alpha=0.7)
plt.title('Stacked Area Plot: Total Rooms and Bedrooms')
plt.xlabel('Index')
plt.ylabel('Count')
plt.legend(loc='upper left')
plt.show()
```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
import pandas as pd
import io
import matplotlib.pyplot as plt

file_name = "housing_california.csv"
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

print(df.head())

# Create a hexbin plot of Latitude and Longitude
plt.figure(figsize=(10, 8))
plt.hexbin(df['longitude'], df['latitude'], gridsize=50, cmap='viridis')
plt.title('Hexbin Plot: Latitude vs. Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Count')
plt.show()
```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
import pandas as pd
import io
import seaborn as sns
import matplotlib.pyplot as plt

file_name = "housing_california.csv"
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

print(df.head())

selected_columns = ['median_income', 'median_house_value', 'population']

# Create a Facet Grid of Scatter Plots
sns.set(style="ticks")
grid = sns.FacetGrid(df, col="ocean_proximity", hue="ocean_proximity", col_wrap=3, height=4)
grid.map(plt.scatter, "median_income", "median_house_value", alpha=0.5)
grid.add_legend()
plt.suptitle('Facet Grid of Scatter Plots: Median Income vs. Median House Value')
plt.show()
```
